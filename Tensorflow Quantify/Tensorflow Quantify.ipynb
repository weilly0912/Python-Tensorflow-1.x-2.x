# Keras to Tensorflow Lite -  (1) Mix_quantify

```python
#混合型量化
import tensorflow as tf
import numpy as np 
def keras_to_tfilte_mix_quantify(kerasModel,TFliteModel):
    converter = tf.compat.v1.lite.TFLiteConverter.from_keras_model_file(kerasModel)
    converter.optimizations = [tf.lite.Optimize.OPTIMIZE_FOR_SIZE]
    tflite_model = converter.convert()
    with open(TFliteModel,'wb') as f:
        f.write(tflite_model)
    print("tranfer done!!")
```

```python
keras_to_tfilte_mix_quantify('./Keras Model/ResNet50.h5', './Keras Model/ResNet50_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/ResNet101.h5', './Keras Model/ResNet101_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/ResNet152.h5', './Keras Model/ResNet152_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/ResNet50V2.h5', './Keras Model/ResNet50V2_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/ResNet101V2.h5', './Keras Model/ResNet101V2_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/ResNet152V2.h5', './Keras Model/ResNet152V2_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/MobileNet.h5', './Keras Model/MobileNet_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/MobileNetV2.h5', './Keras Model/MobileNetV2_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/DenseNet121.h5', './Keras Model/DenseNet121_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/DenseNet169.h5', './Keras Model/DenseNet169_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/InceptionV3.h5', './Keras Model/InceptionV3_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/InceptionResNetV2.h5', './Keras Model/InceptionResNetV2_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/VGG16.h5', './Keras Model/VGG16_Mix.tflite')
keras_to_tfilte_mix_quantify('./Keras Model/VGG19.h5', './Keras Model/VGG19_Mix.tflite')
```

```python
#全整型量化
import tensorflow as tf
import numpy as np 
def keras_to_tfilte_entrie_quantify(kerasModel,TFliteModel):
    converter = tf.compat.v1.lite.TFLiteConverter.from_keras_model_file(kerasModel)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
    converter.inference_input_type  = tf.uint8
    converter.inference_output_type = tf.uint8
    converter.representative_dataset = representative_dataset_gen
    tflite_model = converter.convert()
    with open(TFliteModel,'wb') as f:
        f.write(tflite_model)
    print("tranfer done!!")
    
#converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
```

# Keras to Tensorflow Lite -  (3) Half Float16_quantify

```python
#半精度型量化
import tensorflow as tf
import numpy as np 
def keras_to_tfilte_half_quantify(kerasModel,TFliteModel):
    converter = tf.compat.v1.lite.TFLiteConverter.from_keras_model_file(kerasModel)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    converter.target_spec.supported_types = [tf.float16]
    tflite_model = converter.convert()
    with open(TFliteModel,'wb') as f:
        f.write(tflite_model)
    print("tranfer done!!")
```
